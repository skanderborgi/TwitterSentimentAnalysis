```python
import tweepy
import matplotlib.pyplot as plt
import pandas as pd
from textblob import TextBlob, Word, Blobber


consumer_key = ""
consumer_secret = ""
access_token = ""
access_token_secret = ""




# Creating the authentication object
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
# Setting your access token and secret
auth.set_access_token(access_token, access_token_secret)
# Creating the API object while passing in auth information
api = tweepy.API(auth) 
```

```python
#this will calculate the pourcentage 
def pourcentage(part,whole):
    return 100 * float(part)/float(whole)
```

```python

#Finding Tweets Using a Keyword
api = tweepy.API(auth)

# The search term you want to find
searchTerm = input("Enter keyword/hashtag to search about: ")
nbTerms=input("Enter how many tweets to analyse: ")

# Language code (follows ISO 639-1 standards)
language = "en"

# Calling the user_timeline function with our parameters
results = api.search(q=searchTerm, lang=language,count=nbTerms)
negative=0
positive=0
neutral=0
polarity=0

#dataframe

col_names=['User','Time','Place','Description','Followers','Tweet','Sentiment']
db=pd.DataFrame(columns=col_names)

# foreach through all tweets pulled
for tweet in results:
    # printing the text stored inside the tweet object
    print (tweet.user.screen_name,"Tweeted:",tweet.text,"STATUS:",tweet.id_str)
    if tweet.text.startswith("RT") == True:
        ret=tweet.retweeted_status.id
        print('This tweet is a retweet')
    print("BLALALA",ret)
    print("retweets: "+str(tweet.retweet_count))
    analysis=TextBlob(tweet.text)    
    db = db.append({'User':tweet.user.screen_name, 'Time':tweet.created_at,'Place':tweet.user.location
                   ,'Description':tweet.user.description,'Followers':tweet.user.followers_count,'Tweet':tweet.text,
                   'Sentiment':analysis.sentiment.polarity,'ID':tweet.id_str, 'Retweet':str(ret)}, ignore_index=True)
    #get the sentiment of every tweet
   

    polarity+=analysis.sentiment.polarity
    if(analysis.sentiment.polarity==0):
        neutral+=1
    if(analysis.sentiment.polarity>0.00):
        positive+=1
    if(analysis.sentiment.polarity<0.00):
        negative+=1
positive= pourcentage(positive,nbTerms)
negative=pourcentage(negative,nbTerms)
neutral=pourcentage(neutral,nbTerms)

positive=format(positive,'.2f')
negative=format(negative,'.2f')
neutral=format(neutral,'.2f')
        
print("This is how are people reacting to "+searchTerm+" by analysing " +str(nbTerms)+" Tweets: ")

if (polarity == 0):
    print("Neutral")
if (polarity < 0):
    print("Negative")
if (polarity > 0):
    print("Positive")

labels=["Positive["+str(positive)+"%]","Neutral["+str(neutral)+"%]","Negative["+str(negative)+"%]"]
sizes=[positive,neutral,negative]
colors=['green','yellow','red']
patches, texts=plt.pie(sizes,colors=colors,startangle=90)
plt.legend(patches,labels,loc='best')
plt.axis('equal')
plt.title("This is how are people reacting to "+searchTerm+" by analysing " +str(nbTerms)+" Tweets. ")
plt.tight_layout()
plt.show()
db.to_csv('D:/Data Science/Python for Data Science/Projets/Twitter.csv')
db2=db.iloc[:,7:].copy()
db
```
